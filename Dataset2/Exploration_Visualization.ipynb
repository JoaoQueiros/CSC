## Dataset

# Atributos:
* LandAverageTemperature: Média da temperatura da terra em Celsius
* LandAverageTemperatureUncertainty: Intervalo de 95% de confiança da média
* LandMaxTemperature: Média da temperatura máxima da terra em Celsius
* LandMaxTemperatureUncertainty: Intervalo de 95% de confiança da média da temperatura máxima
* LandMinTemperature: Média da temperatura mínima da terra em Celsius
* LandminTemperatureUncertainty: Intervalo de 95% de confiança da média da temperatura mínima
* LandAndOceanAverageTemperature: Média da temperatura da terra e do oceano em Celsius
* LandAndOceanAverageTemperatureUncertainty: Intervalo de 95% de confiança da média da temperatura da terra e do oceano
